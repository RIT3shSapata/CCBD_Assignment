# Getting overall Statistics

```python
syscalls = dict()
```

### Function that combines system calls in a dictionary

```python
def totalcalls(syscalls, df):
    for i in range(len(df)):
        if df.iloc[i]['syscall'] in syscalls:
            syscalls[df.iloc[i]['syscall']] += df.iloc[i]['seconds']
        else:
            syscalls[df.iloc[i]['syscall']] = df.iloc[i]['seconds']
    return syscalls
```

```python
syscalls = totalcalls(syscalls, browser)
syscalls = totalcalls(syscalls, cat)
syscalls = totalcalls(syscalls, echo)
syscalls = totalcalls(syscalls, browser)
syscalls = totalcalls(syscalls, ls)
syscalls = totalcalls(syscalls, mkdir)
syscalls = totalcalls(syscalls, redis1)
syscalls = totalcalls(syscalls, rmdir)
syscalls
```

### Sorting Dictionary by Value to get top 15 System Calls

```python
sorted_calls = sorted(syscalls, key=syscalls.get, reverse = True)
sorted_calls
```

```python
final_calls = dict()
for i in sorted_calls:
    final_calls[i] = syscalls[i]

final_calls
```

```python
keys = list()
values = list()
for i in final_calls:
    keys.append(i)
    values.append(final_calls[i])
```

# Plotting Graphs

## Top 15 System Calls Across Processes

```python
plt.figure(figsize=(15, 5))
plt.bar(keys[:15], values[:15])

for index, value in enumerate(values[:15]):
    plt.text(index, value, str(value)[:4], ha = "center")
```

### Pie-Chart


```python
def perc(pct):
    return ('%.1f' % pct) if pct > 5 else ''
```

```python
plt.figure()
plt.pie(x = values[:15], labels = keys[:15], labeldistance = None, radius = 3, autopct = perc)
plt.legend(loc = "center", bbox_to_anchor = (1.05, 1))
```
