### Function that combines system calls in a dictionary

```python
def totalcalls(syscalls, df):
    for i in range(len(df)):
        if df.iloc[i]['syscall'] in syscalls:
            syscalls[df.iloc[i]['syscall']] += df.iloc[i]['seconds']
        else:
            syscalls[df.iloc[i]['syscall']] = df.iloc[i]['seconds']
    return syscalls
```
