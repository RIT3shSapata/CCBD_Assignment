# Getting overall Statistics

### Function that combines system calls in a dictionary by time

```python
def totalcalls(syscalls, df):
    for i in range(len(df)):
        if df.iloc[i]['syscall'] in syscalls:
            syscalls[df.iloc[i]['syscall']] += df.iloc[i]['seconds']
        else:
            syscalls[df.iloc[i]['syscall']] = df.iloc[i]['seconds']
    return syscalls
```

### Function that combines system calls in a dictionary by number

```python
def calls_by_num(callsnum, df):
    for i in range(len(df)):
        if df.iloc[i]['syscall'] in callsnum:
            callsnum[df.iloc[i]['syscall']] += df.iloc[i]['calls']
        else:
            callsnum[df.iloc[i]['syscall']] = df.iloc[i]['calls']
    return callsnum
```

### Function that combines system calls in a dictionary by errors

```python
def calls_error(errorcalls, df):
    for i in range(len(df)):
        if df.iloc[i]['syscall'] in errorcalls:
            errorcalls[df.iloc[i]['syscall']] += df.iloc[i]['errors']
        else:
            errorcalls[df.iloc[i]['syscall']] = df.iloc[i]['errors']
    return errorcalls
```
